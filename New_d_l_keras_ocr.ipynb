```python
import keras_ocr
import matplotlib.pyplot as plt
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
images = [
    keras_ocr.tools.read(img) for img in ['/content/dl-fr.jpg'
    ]
]
```

```python
len(images)
```

```python
plt.figure(figsize = (10,20))
plt.imshow(images[0])
```

```python
prediction_groups = pipeline.recognize(images)
```

```python
prediction_groups
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image,
                                    predictions=predictions,
                                    ax=ax)
```

```python
import keras_ocr
import matplotlib.pyplot as plt

pipeline = keras_ocr.pipeline.Pipeline()

image_paths = ['/content/dl-fr.jpg']

prediction_groups = pipeline.recognize(image_paths)


for image_path, predictions in zip(image_paths, prediction_groups):
    print(f"Text in '{image_path}':")
    for prediction in predictions:
        print(prediction[0])
    print("\n")

plt.close('all')
```
